# linear regression



Linear  Regression is a statistical method  used to model the relationship between a
dependent variable(target) and one more independent variables(feature) .

the goal is too find the linear equation that best predicts the target variable from 
the feature variables

Equation: The mathematical representation of a linear regression model is (y = a_0 + a_1x + \epsilon), where:

(y) is the dependent variable (target).

(x) is the independent variable (feature).

(a_0) is the intercept of the line, giving an additional degree of freedom.

(a_1) is the linear regression coefficient, acting as a scale factor for each input value

(\epsilon) represents the random error.

#Example

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
```

```python
# Example data

data = {
    'Size' : [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400],
    'Price' : [300000, 320000, 340000, 360000, 380000, 400000, 420000, 440000, 460000, 480000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#Independent variable (feature) and Dependent variable (target)

X = df[['Size']]
y = df['Price']
```

```python
#splitting th data into the training and testing sets

X_train, X_test, y_train, y_test = train_test_split( X, y,test_size = 0.2, random_state = 0)
```

```python
#Creating the training and the linear regression model

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#Evaluating the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error : {mse}")
print(f"R-Squared : {r2}")
```

```python
#Plotting the  the result

plt.scatter(X, y, color='blue')   #Original data points

plt.plot(X_test, y_pred, color='red',linewidth = 2)  #Regression line

plt.xlabel('Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('Linear Regression : House Price vs Size')
plt.show()
```
