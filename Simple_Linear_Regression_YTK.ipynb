```python
# importing libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Loading  data

data_url = "http://lib.stat.cmu.edu/datasets/boston"

df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([df.values[::2, :], df.values[1::2, :2]])
target = df.values[1::2, 2]

import requests

response = requests.get(data_url)
lines = response.text.splitlines()

feature_names = []
for i in range(1, 22):
    row = lines[i]
    words = row.split()
    if words:
        feature_name = words[0]
        feature_names.append(feature_name)

feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
```

```python
data
```

```python
target
```

```python
feature_names
```

```python
#Convertion of data into DataFrame

df = pd.DataFrame(data, columns=feature_names)
```

```python
df
```

```python
# Dividing Dependent and independent values

X = df
y = target
```

```python
y
```

```python
# Train test split(Splitting the data)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
# Standardizing the dataset

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#model building

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
#Cross validation

cross_val_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
```

```python
cross_val_scores
```

```python
np.mean(cross_val_scores)
```

```python
#prediction
model_pred = model.predict(X_test)
```

```python
model_pred
```

```python
#Ploting differnece between actual values and predicted

sns.distplot(model_pred - y_test)
```

```python
#Regression Score Function

score = r2_score(model_pred, y_test)
```
